```python
tipSayisi=0
toplamAmpulTipi=0
while tipSayisi < 2:
      tipSayisi = int(input("Ampul Tip sayısını giriniz:"), 10)

ampul             = [None] * tipSayisi #Ampul Tip Sayısı "BOŞ" tanımlanıyor
ampulTipiOran     = [None] * tipSayisi #ampul tipi oranları listesi "BOŞ" tanımlanıyor
tipOlmaOlasiligi  = [None] * tipSayisi #Her bir ampul tipindeki yüzde oranı listesi "BOŞ" tanımlanıyor

##Her bir Tip Ampule Ait Olma Olasılığı
for i in range(tipSayisi):
    ampul[i] = input(str(i+1) + ". Tip Ampulun Kutu İçerisindeki Olma Oranını % olarak giriniz:")
```

image.png

```python
##Her bir Tip Ampule 100 Saat Üzerinde Işık Verme Olasılığı
for i in range(tipSayisi):
    ampulTipiOran[i] = input(str(i+1) + ". Tip Ampulun 100 Saat Üzerinde Yanma Oranını % olarak giriniz:")
```

image.png

```python
##Seçilen Ampul Tipinin hangi oranda 100 Saaatin üzerinde olabilir?
print("Tipi ve 100 saat üzerinde yanma olasılığı bilindiğine göre;")
yananAmpulTipi = input("1 - " + str(tipSayisi) + ". tip ampul arasından hangi tip olduğu hesaplanmalı?")

for i in range(tipSayisi):
    tipOlmaOlasiligi[i] = (int(ampul[i], 10) / 100) * (int(ampulTipiOran[i], 10) / 100)
    toplamAmpulTipi += tipOlmaOlasiligi[i]
    
secim=int(yananAmpulTipi)-1
olasilik = tipOlmaOlasiligi[secim] / toplamAmpulTipi

print("-----------------------------------")
print("Ampullerin Kutu İçerisindeki Bulunma Oranları")
print(ampul)
print("100 Saat Üzerinde Yanma Olasılığı")
print(ampulTipiOran)
print("Seçilen Ampul Tipi Olasılğı")
print(tipOlmaOlasiligi)
print("Tüm ampul tiplerinin yanma olasılığı")
print(toplamAmpulTipi)

print("Seçili ampulün olma olasılığı :",olasilik)
```
